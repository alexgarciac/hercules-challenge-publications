```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
```

# Example Data

```python
mean_test_score = [0.73899224, 0.7443584 , 0.75657843, 0.7443584 , 0.74415484,
                   0.75657843, 0.74704918, 0.7643855 , 0.7643855 , 0.73899224,
                   0.75657843, 0.7643855 , 0.7443584 , 0.74704918, 0.73899224]

params = [(15, 0.9),(20, 0.7),(30, 0.5),(20, 0.5),(25, 0.7),(30, 0.7),
          (10, 0.5),(5, 0.5),(5, 0.7),(15, 0.5),(30, 0.9),(5, 0.9),
          (20, 0.9),(10, 0.9),(15, 0.7)]
```

# Group Values
Group the values based on their learning decay

```python
learning_decay_05 = []
learning_decay_07 = []
learning_decay_09 = []

index = 0
for i in params:
    n_components = i[0]
    learning_decay = i[1]
    test_score = mean_test_score[index]
    if(learning_decay==0.5):
        learning_decay_05.append((n_components,test_score))
    if(learning_decay==0.7):
        learning_decay_07.append((n_components,test_score))
    if(learning_decay==0.9):
        learning_decay_09.append((n_components,test_score))

    index+=1
```

# Order Values
Orther them according to their n_components value

```python
def getKey(item):
       return item[0]

learning_decay_05 = sorted(learning_decay_05, key=getKey);
learning_decay_07 = sorted(learning_decay_07, key=getKey);
learning_decay_09 = sorted(learning_decay_09, key=getKey);
```

# Draw the plot

```python
output_notebook()

p = figure(
    title='Test Score results',
    x_axis_label='number of topics',
    y_axis_label='coherence',
    plot_width=1000, 
    plot_height=400
)

def get_values_by_index(learning_decay,value_index):
       values = []
       for l in learning_decay:
              values.append(l[value_index])
       return values

p.line(get_values_by_index(learning_decay_05,0),
       get_values_by_index(learning_decay_05,1),
       legend_label='0.5',color="green",line_width=4)

p.line(get_values_by_index(learning_decay_07,0),
       get_values_by_index(learning_decay_07,1),
       legend_label='0.7',color="red",line_width=4)

p.line(get_values_by_index(learning_decay_09,0),
       get_values_by_index(learning_decay_09,1),
       legend_label='0.9',color="blue",line_width=4)

show(p)
```
