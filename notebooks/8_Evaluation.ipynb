```python
from nltk.corpus import wordnet as wn

def _get_synset(word):
    try:
        word = '_'.join(word.split(' '))
        return wn.synsets(word)[0]
    except IndexError:
        return None
```

```python
from src.utils import load_object

final_pipe = load_object(os.path.join(NOTEBOOK_7_RESULTS_DIR, 'final_pipe.pkl'))
```

```python
articles_keys = final_articles_categories.keys()
X = categories_df.set_index('id', inplace=False).loc[articles_keys]['text_cleaned'].values
```

```python
y_base = final_articles_categories.values()
y_pred = final_pipe.transform(X)
```

```python
y_pred = [[topic[0] for topic in doc] for doc in y_pred]
```

```python
from nltk.corpus.reader.wordnet import WordNetError

def get_score(topics_base, topics_pred, similarity_func):
    sim_measures = []
    for topic_p in topics_pred:
        p_synset = _get_synset(topic_p)
        if p_synset is None:
            continue

        for topic_b in topics_base:
            b_synset = _get_synset(topic_b)
            if b_synset is None:
                continue
            
            try:
                similarity = getattr(p_synset, similarity_func)(b_synset)
                sim_measures.append(similarity)
            except WordNetError:
                # error when comparing synsets with different pos
                continue
    return {
        'max similarity': np.max(sim_measures),
        'min similarity': np.min(sim_measures),
        'mean similarity': np.mean(sim_measures),
        'median similarity': np.median(sim_measures)
    }
```

## Similarity

```python
scores = [get_score(y_b, y_p, 'lch_similarity')
          for y_b, y_p in zip(y_base, y_pred)]
scores[:5]
```
